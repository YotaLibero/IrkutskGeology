# Попытка создания `огибающей сейсмограммы` (`Seismogram Envelopes`)

```python
import obspy.signal

# Фильтрация потокового объекта
st_filt_0 = st.copy()
st_filt_0.filter('bandpass', freqmin=1, freqmax=3, corners=2, zerophase=True)

# Огибание (обёртка) отфильтрованных данных
data_envelope_0 = obspy.signal.filter.envelope(st_filt_0[0].data)
data_envelope_1 = obspy.signal.filter.envelope(st_filt_0[1].data)
data_envelope_2 = obspy.signal.filter.envelope(st_filt_0[2].data)
data_envelope_3 = obspy.signal.filter.envelope(st_filt_0[3].data)

# Построение графика (с помощью matplotlib)
t_0 = np.arange(0, st[0].stats.npts / st[0].stats.sampling_rate, 1 / st[0].stats.sampling_rate)
t_1 = np.arange(0, st[1].stats.npts / st[1].stats.sampling_rate, 1 / st[1].stats.sampling_rate)
t_2 = np.arange(0, st[2].stats.npts / st[2].stats.sampling_rate, 1 / st[2].stats.sampling_rate)
t_3 = np.arange(0, st[3].stats.npts / st[3].stats.sampling_rate, 1 / st[3].stats.sampling_rate)

plt.figure(figsize=(20, 16))

plt.subplot(4,1,1)
plt.plot(t_0, st_filt_0[0].data, 'k')
plt.plot(t_0, data_envelope_0, 'k:')
plt.title(st[0].stats.starttime)
plt.ylabel('Filtered Data w/ Envelope')
# plt.xlabel('Time [s]')
plt.xlim(80, 90)

plt.subplot(4,1,2)
plt.plot(t_1, st_filt_0[1].data, 'k')
plt.plot(t_1, data_envelope_1, 'k:')
plt.title(st[1].stats.starttime)
plt.ylabel('Filtered Data w/ Envelope')
# plt.xlabel('Time [s]')
plt.xlim(80, 90)

plt.subplot(4,1,3)
plt.plot(t_2, st_filt_0[2].data, 'k')
plt.plot(t_2, data_envelope_2, 'k:')
plt.title(st[2].stats.starttime)
plt.ylabel('Filtered Data w/ Envelope')
# plt.xlabel('Time [s]')
plt.xlim(80, 90)

plt.subplot(4,1,4)
plt.plot(t_3, st_filt_0[3].data, 'k')
plt.plot(t_3, data_envelope_3, 'k:')
plt.title(st[3].stats.starttime)
plt.ylabel('Filtered Data w/ Envelope')
plt.xlabel('Time [s]')
plt.xlim(80, 90)

plt.show()
```

# ЭКСПЕРИМЕНТ ПО ПОСТРОЕНИЮ НЕПРЕРЫВНОГО ВЕЙВЛЕТ–ПРЕОБРАЗОВАНИЯ

[Ссылка на документацию](https://docs.obspy.org/tutorial/code_snippets/continuous_wavelet_transform.html)




```python
from obspy.imaging.cm import obspy_sequential
from obspy.signal.tf_misfit import cwt

TR = st[0]
NPTS = TR.stats.npts
DT = TR.stats.delta
T = np.linspace(0, DT * NPTS, NPTS)
F_MIN = .01
F_MAX = 10

scalogram = cwt(TR.data, DT, 8, F_MIN, F_MAX)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(1,1,1)

x, y = np.meshgrid(
    t,
    np.logspace(np.log10(F_MIN), np.log10(F_MAX), scalogram.shape[0])
)

ax.pcolormesh(x, y, np.abs(scalogram), cmap=obspy_sequential)
ax.set_xlabel('Time after %s [s]' % TR.stats.starttime)
ax.set_ylabel('Frequency [Hz]')
ax.set_yscale('log')
ax.set_ylim(F_MIN, F_MAX)

plt.show()
```
